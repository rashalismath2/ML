<h3><b>Question:</b>Participant dencity by occupation and country</h3>

```python
pd.crosstab(df['Country'], df['Occupation']).plot(kind='bar', stacked=True)
plt.ylabel('Count')
plt.title('Occupation Distribution by Country')
plt.show()
```

```python
df_us=df[df['Country'] == 'United States']
```

```python
df_us.info()
```

<h3><b>Question:</b>Bar chart for occupations in US</h3>

```python
df_us['Occupation'].value_counts().plot(kind='bar')

plt.title('Count of Occupations')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.show()
```

```python
df_us['Occupation'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Occupation Distribution')
plt.show()
```

<h3><b>Question:</b>Work interest vs occupation</h3>

```python
table = pd.crosstab(df_us['Occupation'], df_us['Work_Interest'], normalize='index') * 100
print(table)
```

```python
table.plot(kind='bar', stacked=True)
plt.ylabel("Percentage")
plt.title("Work Interest by Occupation")
plt.show()
```

<h3><b>Question:</b>Work interest vs occupation vs gender</h3>

```python
counts = df_us.groupby(['Occupation', 'Work_Interest', 'Gender']).size().reset_index(name='count')
print(counts)
```

```python
sns.catplot(
    data=df_us,
    x='Occupation',
    hue='Work_Interest',
    col='Gender',
    kind='count',
    height=5,
    aspect=1.2
)

plt.show()
```

<h3><b>Question:</b>Work interest vs occupation vs gender in each occupation</h3>

```python
sns.catplot(
    data=df_us,
    x="Occupation",
    hue="Gender",
    row="Work_Interest",
    kind="count",
    height=4,
    aspect=2, palette={"Male": "blue", "Female": "pink"}
)

plt.xticks(rotation=45)
plt.show()
```
